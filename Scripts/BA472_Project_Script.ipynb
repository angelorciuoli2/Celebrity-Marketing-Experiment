#Respondent's Estimated Prices Analysis

```python
# Product 1: Hoodie

df = pd.read_csv('BA472_Survey_Results_Cleaned.csv')
hoodie_df = df[['Treatment', 'Hoodie_Guessed_Price']]
atehat_hoodie = hoodie_df.loc[df['Treatment'] == 1, 'Hoodie_Guessed_Price'].mean() - hoodie_df.loc[df['Treatment'] == 0, 'Hoodie_Guessed_Price'].mean()
print("ATE for Respondent's Estimated Price on Hoodie: $", round(atehat_hoodie,2))

tstat, pvalue, df = ttest_ind(hoodie_df.loc[hoodie_df['Treatment'] == 1, 'Hoodie_Guessed_Price'],
                              hoodie_df.loc[hoodie_df['Treatment'] == 0, 'Hoodie_Guessed_Price'],
                              alternative = 'two-sided', usevar = 'unequal', value = 0)

print(f"Hoodie t-score (t): {tstat:.4f}")
print(f"Hoodie P-value (p): {pvalue:.4e}")
print(f"Should we reject the null hypothesis that celebrity marketing has no effect on respondent's estimated price on Hoodie? {pvalue < 0.05}")

hoodie_ttest_result = ttest(hoodie_df.loc[hoodie_df['Treatment'] == 1, 'Hoodie_Guessed_Price'], hoodie_df.loc[hoodie_df['Treatment'] == 0, 'Hoodie_Guessed_Price'])
hoodie_ttest_result
```

```python
# CATE
df = pd.read_csv('BA472_Survey_Results_Cleaned.csv')
dummy_df = pd.get_dummies(df['Sex']).astype(int)
df = pd.concat([df, dummy_df], axis=1)
df = df.drop('Sex', axis=1)
df = df.drop('Other', axis=1)
products = ['Hoodie','Bag','Loafer','Glasses','Jacket','Mascara']
for i in products:
  formula_intent_self = f"{i}_Purchase_Intent_Self ~ Treatment*Male"
  formula_intent_friend = f"{i}_Purchase_Intent_Friend ~ Treatment*Male"
  formula_guessed_price = f"{i}_Guessed_Price ~ Treatment*Male"

  model_intent_self = smf.ols(formula_intent_self, data=df).fit(cov_type='HC1')
  model_intent_friend = smf.ols(formula_intent_friend, data=df).fit(cov_type='HC1')
  model_guessed_price = smf.ols(formula_guessed_price, data=df).fit(cov_type='HC1')

  print(f"**Product: {i}**")
  print("Regression Results:")
  print('PI_self*Male:',model_intent_self.params['Treatment']+model_intent_self.params['Treatment:Male'],'|p-val:',(model_intent_self.pvalues['Treatment']+model_intent_self.pvalues['Treatment:Male'])/2,
        '\nPI_friend*Male:',model_intent_friend.params['Treatment']+model_intent_friend.params['Treatment:Male'],'|p-val:',(model_intent_friend.pvalues['Treatment']+model_intent_friend.pvalues['Treatment:Male'])/2,
        '\nPP*Male:',model_guessed_price.params['Treatment']+model_guessed_price.params['Treatment:Male'],'|p-val:',(model_guessed_price.pvalues['Treatment']+model_guessed_price.pvalues['Treatment:Male'])/2)
```
