**Varible Definitions**

| Variable          | Variable Definitions                                                                                                                               |
|-------------------|----------------------------------------------------------------------------------------------------------------------------------------------------------------------------|
| `RespondId`   |  automatically generated ID for survey participant                                                                                                 |
| `Treatment`   | treatment = 1 if respondent took treatment survey                                                                                                                      |
| `Age`   | Age                                                                                                          |
| `Sex`   |  Male, Female, or Other                                                                                                 |
| `Respondent_Budget` | Respondents average monthly spending on similar products (Categorical variable, with 5 unique classes)                                                                                                   |
| `Product_Purchase_Intent_Self` | A scale of 1 to 5, how likely the respondent would purchase this product for themselves (6 of these variables for 6 different products)                                                                                                    |
| `Product_Purchase_Intent_Friend` | A scale of 1 to 5, how likely the respondent would purchase this product for someone they know, not neccessarily a friend (6 of these variables for 6 different products)                                                                                                     |
| `Product_Guessed_Price` | Respondent's guess of how much the product is sold for. (6 of these variables for 6 different products)                                                                                                     |
| `Product_Celeb` | 1 if the treatment respondent knows who the celebrity is, 0 if they don't know, and NaN if the respondent took the control survey. (6 of these variables for 6 different products)                                                                                                    |

```python
df = pd.read_csv('BA472 Project Survey_April 30, 2024_19.00.csv')
df
```

```python
# Removing initial rows (Survey test)
df = df.dropna(subset=['IPAddress'])
df = df.drop(df.index[:2])
df = df[df['Finished'] != 'False']

# Dropping unneccessary columns
df = df.drop('StartDate', axis=1)
df = df.drop('EndDate', axis=1)
df = df.drop('Status', axis=1)
df = df.drop('Progress', axis=1)
df = df.drop('Finished', axis=1)
df = df.drop('RecordedDate', axis=1)
df = df.drop('RecipientLastName', axis=1)
df = df.drop('RecipientFirstName', axis=1)
df = df.drop('RecipientEmail', axis=1)
df = df.drop('ExternalReference', axis=1)
df = df.drop('DistributionChannel', axis=1)
df = df.drop('UserLanguage', axis=1)
df = df.drop('LocationLatitude', axis=1)
df = df.drop('LocationLongitude', axis=1)
df = df.drop('IPAddress', axis=1)
df = df.drop('Duration (in seconds)', axis=1)

#Initializing Treatment column
df['Treatment'] = np.NaN
df
```

```python
# Making treatment = 1 if all treatment questions are not null

treatment_columns = ['T1.1_8', 'T1.2_8', 'T1.3', 'T1.4', 'T2.1_8', 'T2.2_8', 'T2.3', 'T2.4', 'T3.1_8',
                    'T3.2_8', 'T3.3', 'T3.4', 'T4.1_8', 'T4.2_8', 'T4.3', 'T4.4', 'T5.1_8', 'T5.2_8',
                    'T5.3', 'T5.4', 'T6.1_8', 'T6.2_8', 'T6.3', 'T6.4']

# Check if all specified columns are not null for each row
treatment_not_null = df[treatment_columns].notnull().all(axis=1)

# Amend the 'Treatment' column to 1 where all specified columns are not null
df.loc[treatment_not_null, 'Treatment'] = 1

# Making control = 0 if all control questions are not null

control_columns = ['C1.1_1', 'C1.2_8', 'C1.3', 'C2.1_8', 'C2.2_8', 'C2.3', 'C3.1_8', 'C3.2_8',
                   'C3.3', 'C4.1_8', 'C4.2_8', 'C4.3', 'C5.1_8', 'C5.2_8', 'C5.3', 'C6.1_8',
                   'C6.2_8', 'C6.3']

# Check if all specified columns are not null for each row
control_not_null = df[control_columns].notnull().all(axis=1)

# Amend the 'Treatment' column to 0 where all specified columns are not null
df.loc[control_not_null, 'Treatment'] = 0

df = df.dropna(subset=['Treatment'])
df['Treatment'] = df['Treatment'].astype(int)
```
